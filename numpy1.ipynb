#load in numpy


```python
a = np.array([1,2,3])
print(a)
#get dimension
print(a.ndim)
#get shape
print(a.shape)
#get type
print(a.dtype)
#get size
print(a.itemsize)
#get totelsize
a.nbytes
```

```python
b = np.array([[1.0,3.2,4.3],[3.6,7.3,8.2]],dtype='int16')
print(b)
#get dimension
print(b.ndim)
#get shape
print(b.shape)
#get type
print(b.dtype)
#get size
print(b.itemsize)
#get total size
b.nbytes
```
